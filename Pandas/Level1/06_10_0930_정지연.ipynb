# 06. 상위 n개 레코드

```python
import pandas as pd
animal_df = pd.read_csv('../datasets/ANIMAL_INS.csv')
```

```python
animal_df
```

```python
animal_df[['NAME', 'DATETIME']].sort_values('DATETIME')   
# DF를 슬라이싱 할거면, SORT하고자 하는 값이 앞에 꼭 들어가야함. 
# EX. " animal_df[['NAME']].sort_values('DATETIME') "는 이미 슬라이싱 된 DATETIME이 없는 DF이기 때문에, DATETIME 기준으로 sort_vlaues를 할 수 없음
```

```python
animal_df.sort_values('DATETIME')[['NAME']]
# 이렇게 하면 'DATETIME'으로 이미 정렬이 된 채로 DF가 나오고, 그 다음 'NAME'을 슬라이싱하기 때문에, 값이 나올 수 있음.
```

```python
# 동물 보호소에 가장 먼저 들어온 동물의 이름을 조회
animal_df.sort_values('DATETIME')[['NAME']].iloc[0]
```

```python
animal_df['DATETIME'] = pd.to_datetime(animal_df['DATETIME'], format='mixed')
```
